```python
import numpy as np
import matplotlib.pyplot as plt

# Fórmulas corregidas
def esfuerzo(S):
    return 8 * S**0.95  # E = 8 × S^0.95

def tiempo_calendario(E):
    return 2.4 * E**0.33  # td = 2.4 × E^0.33

# Gráfico 1: E vs S (S ∈ [0, 10000])
S_values = np.linspace(0, 10000, 500)
E_values = esfuerzo(S_values)

plt.figure(figsize=(10, 5))
plt.plot(S_values, E_values, label='E = 8 × S^0.95')
plt.title('Esfuerzo (E) en función del Tamaño (S)')
plt.xlabel('Tamaño del proyecto (S)')
plt.ylabel('Esfuerzo (E)')
plt.grid(True)
plt.legend()
plt.show()

# Gráfico 2: td vs E (E ∈ [1, 500])
E_range = np.linspace(1, 500, 500)
td_values = tiempo_calendario(E_range)

plt.figure(figsize=(10, 5))
plt.plot(E_range, td_values, label='td = 2.4 × E^0.33', color='orange')
plt.title('Tiempo Calendario (td) en función del Esfuerzo (E)')
plt.xlabel('Esfuerzo (E)')
plt.ylabel('Tiempo Calendario (td)')
plt.grid(True)
plt.legend()
plt.show()
```
